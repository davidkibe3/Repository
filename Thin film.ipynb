```python
import numpy as np
from matplotlib import pylab as plt
N=100
M=200
mu=1
dt=0.1
dy=0.1
rho=1000
c=1
k=1
ro=1
To=30
ki=1
b=1
a=0.8*b
Pr=mu*c/k
R=16*ro*To**3/(3*k*ki)
x=np.linspace(0,1,N)
y=np.linspace(0,1,N)
u=np.zeros(N)
usol=np.zeros(N)
T=To*np.ones(N)
Tsol=np.zeros(N)
Ts=1
U=np.ones(1)
t = 1 #time loop start at t=0
while t <=M: #Loop for time 
    t = t + 1
    U=b*x/(1-a*t)
    #=To-(Tref*(b*x**2)/(2*mu))*(1-a*t)**(-3/2)
    for j in range(1,N-1):
        usol[j]=u[j]+(mu*dt/(rho*dy**2))*(u[j+1]-2*u[j]+u[j-1])
        Tsol[j]=(dt/dy**2)*(mu/(rho*Pr+R*k/(rho*c)))*(T[j+1]-2*T[j]+T[j-1])+T[j]
    u[0]=U[0]
    T[0]=Ts
    u[-1]=u[-2]
    T[-1]=T[-2]
    u[0:]=usol
    T[0:]=Tsol
#print(U)
```

```python
import numpy as np
from matplotlib import pylab as plt
N=200
M=5000
L=1
g=9.8
dt=0.01
dx=0.1
rho=1300
theta=np.pi/12
To=0
n=1000
x=np.linspace(0,L,N)
T=np.zeros(N)
u=10*np.ones(N)
usol=np.zeros(N)
Tsol=np.zeros(N)
count = 1 #time loop start at t=0
while count <=M: #Loop for time 
    count = count + 1
    for j in range (1,N-1):
        Tsol[j]=To+(n/(2*dx))*(u[j+1]-u[j-1])
        usol[j]=(dt/(2*dx*rho))*(T[j+1]-T[j-1])+g*dt*np.sin(theta)+u[j]#+(dt/dx**2)*(u[j+1]-2*u[j]+u[j-1])
    #u[0]=0
    #u[-1]=0
    T[1]=T[0]
    T[-1]=T[-2]
    u[0:]=usol
    T[0:]=Tsol
```

```python
plt.plot(x,u,'b')
plt.xlabel(r'$y$')
plt.ylabel(r'u')
plt.grid()
plt.show()

plt.plot(x,T,'r')
plt.xlabel(r'$y$')
plt.ylabel(r'$\tau$')
plt.grid()
plt.show()
```
